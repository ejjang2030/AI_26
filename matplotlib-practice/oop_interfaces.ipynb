```python
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
ys = [np.sin(x), np.cos(x), 2 * np.sin(x), 2 * np.cos(x)]

fig, axes = plt.subplots(4, 1, figsize=(5, 8))

for ax, y in zip(axes, ys):
    ax.plot(x, y)

plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
ys = [np.sin(x), np.cos(x), 2 * np.sin(x), 2 * np.cos(x)]

fig, axes = plt.subplots(4, 1, figsize=(5, 8))

for ax_idx, ax in enumerate(axes):
    ax.plot(x, ys[ax_idx])
    ax.tick_params(labelsize=13)
fig.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
ys = [np.sin(x), np.cos(x), 2 * np.sin(x), 2 * np.cos(x)]

fig, axes = plt.subplots(4, 1, figsize=(5, 8))

for ax_idx, ax in enumerate(axes):
    ax.plot(x, ys[ax_idx])
    ax.tick_params(labelsize=13)
    ax.set_xlabel('x', fontsize=15)
    ax.set_ylabel('y', fontsize=15)
fig.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
ys = [np.sin(x), np.cos(x), 2 * np.sin(x), 2 * np.cos(x)]
titles = [r'$y = \sin(x)$', r'$y = \cos(x)$', r'$y = 2\sin(x)$', r'$y = 2\cos(x)$']

fig, axes = plt.subplots(4, 1, figsize=(5, 8))

for ax_idx, ax in enumerate(axes):
    ax.plot(x, ys[ax_idx])
    ax.set_title(titles[ax_idx], fontsize=15)
    ax.tick_params(labelsize=13)
    ax.set_xlabel('x', fontsize=15)
    ax.set_ylabel('y', fontsize=15)
fig.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
ys = [np.sin(x), np.cos(x), 2 * np.sin(x), 2 * np.cos(x)]
titles = [r'$y = \sin(x)$', r'$y = \cos(x)$', r'$y = 2\sin(x)$', r'$y = 2\cos(x)$']

fig, axes = plt.subplots(4, 1, figsize=(5, 8))

for ax_idx, ax in enumerate(axes):
    ax.plot(x, ys[ax_idx])
    ax.set_title(titles[ax_idx], fontsize=15)
    ax.tick_params(labelsize=13)
    if ax_idx == len(axes) - 1:
        ax.set_xlabel('x', fontsize=15)
    ax.set_ylabel('y', fontsize=15)
fig.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
ys = [np.sin(x), np.cos(x), 2 * np.sin(x), 2 * np.cos(x)]
titles = [r'$y = \sin(x)$', r'$y = \cos(x)$', r'$y = 2\sin(x)$', r'$y = 2\cos(x)$']

fig, axes = plt.subplots(2, 2, figsize=(5, 8))

for ax_idx, ax in enumerate(axes.flat):
    ax.plot(x, ys[ax_idx])
    ax.set_title(titles[ax_idx], fontsize=15)
    ax.tick_params(labelsize=13)
    if ax_idx in [2, 3]:
        ax.set_xlabel('x', fontsize=15)
    if ax_idx in [0, 2]:
        ax.set_ylabel('y', fontsize=15)
fig.tight_layout()
plt.show()
```
