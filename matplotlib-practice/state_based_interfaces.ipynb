```python
import numpy as np
```

```python
x = np.linspace(-3, 3, 100)
y1 = x ** 2
y2 = (x - 1) ** 2
y3 = (x + 1) ** 2

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y1 = x ** 2
y2 = (x - 1) ** 2
y3 = (x + 1) ** 2

plt.plot(x, y1, label='$y = x^2$')
plt.plot(x, y2, label='$y = (x - 1)^2$')
plt.plot(x, y3, label='$y = (x + 1)^2$')
plt.legend(fontsize=12)
plt.show()
```

```python
std = 0.5
rng = np.random.default_rng(0)
```

```python
x1_1 = rng.normal(loc=1, scale=std, size=100)
y1_1 = rng.normal(loc=1, scale=std, size=100)
```

```python
x1_2 = rng.normal(loc=-1, scale=std, size=100)
y1_2 = rng.normal(loc=-1, scale=std, size=100)
```

```python
x1 = np.concatenate([x1_1, x1_2])
y1 = np.concatenate([y1_1, y1_2])
```

```python
x2_1 = rng.normal(loc=-1, scale=std, size=100)
y2_1 = rng.normal(loc=1, scale=std, size=100)
```

```python
x2_2 = rng.normal(loc=1, scale=std, size=100)
y2_2 = rng.normal(loc=-1, scale=std, size=100)
```

```python
x2 = np.concatenate([x2_1, x2_2])
y2 = np.concatenate([y2_1, y2_2])
```

```python
plt.scatter(x1, y1, color='b', alpha=0.5, label='positive data')
plt.scatter(x2, y2, color='r', alpha=0.5, label='negative data')
plt.legend(fontsize=12)

plt.show()
```
