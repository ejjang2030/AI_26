```python
import numpy as np

# 1. 샘플 데이터 생성 (2학급, 3학생, 3과목)
# 점수는 50점에서 100점 사이의 랜덤한 정수로 생성합니다.
np.random.seed(42)  # 결과 재현을 위한 시드 설정
scores = np.random.randint(50, 101, size=(2, 3, 3))

print("전체 성적 데이터 (3차원 배열):")
print(scores)
print("-" * 30)

# 2. 특정 데이터 접근 예시
class1_student2_english = scores[0, 1, 1]  # 1반(0), 2번째 학생(1), 영어(1)
print(f"1반 2번 학생의 영어 점수: {class1_student2_english}")

# 3. 데이터 분석 활용
# 각 학급별 전 과목 평균
class_avg = scores.mean(axis=(1, 2))
print(f"1반 평균: {class_avg[0]:.2f}, 2반 평균: {class_avg[1]:.2f}")

# 각 과목별 전체 평균 (국, 영, 수)
subject_avg = scores.mean(axis=(0, 1))
print(f"국어 평균: {subject_avg[0]:.2f}, 영어 평균: {subject_avg[1]:.2f}, 수학 평균: {subject_avg[2]:.2f}")
```

```python
# 연습문제
# 1번
customer = np.array([[150, 25, 0], [200, 30, 0], [50, 10, 1], [300, 45, 0]])
```

```python
print(customer)
```

```python
combined_data = np.vstack((customer, np.array([[120, 20, 0], [180, 35, 1], [75, 5, 0], [160, 25, 0]])))
print(combined_data)
```

```python
total_comment_length, total_likes, _ = combined_data.sum(axis=0)
```

```python
total_comment_length, total_likes
```

```python
average_comment_length, average_likes, _ = combined_data.mean(axis=0)
average_comment_length, average_likes
```

```python
print(f'댓글 길이 총합: {total_comment_length}')
print(f'좋아요 수 총합: {total_likes}')
print(f'댓글 길이 평균: {average_comment_length}')
print(f'좋아요 수 평균: {average_likes}')
```

```python
np.vsplit(combined_data, 4)
```

```python
selected_customers = combined_data[[0, 2, 4]]
print(selected_customers)
```

```python
combined_data.T
```

```python
new_customers = np.zeros((2, 3), dtype=np.int64)
print(new_customers)
```

```python
print(combined_data)
```

```python
a = combined_data[:, 0] >= 100
b = combined_data[:, 1] >= 20
c = combined_data[:, 2] == 0
print(a)
print(b)
print(c)
```

```python
np.all((a, b, c), axis=0)
```

```python
selected_customers = combined_data[np.all((a, b, c), axis=0)]
print(selected_customers)
```
