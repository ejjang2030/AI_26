```python
def conv2d_multi_channel(image, kernel):
    # image: (H, W, C)
    # kernel: (KH, KW, C)
    h, w, c, _ = image.shape
    kh, kw, _, _ = kernel.shape

    # 출력 배열 초기화
    output = np.zeros((h - kh + 1, w - kw + 1))

    # 각 채널별로 2D 합성곱 수행 후 합산
    for i in range(c):
        output += scipy.signal.convolve2d(image[:, :, i], kernel[:, :, i], mode='valid')

    return output
```

```python
sample_image = train_input[0]
sample_image = sample_image.reshape(28, 28)
kernel_weights, kernel_bias = conv1.weights
kernel_weights = np.array(kernel_weights).reshape(32, 3, 3)
print(sample_image.shape, kernel_weights.shape)
arr = []
img_w, img_h = sample_image.shape
n, k_w, k_h = kernel_weights.shape

nn = []
for k in range(n):
    ss = []
    for i in range(img_w - k_w + 1):
        for j in range(img_h - k_h + 1):
            grid = sample_image[i:i + k_w,j:j + k_h]
            ss.append(np.sum(grid * kernel_weights[k]))
    ss = np.array(ss).reshape(img_w - k_w + 1, img_h - k_h + 1)
    nn.append(ss)

nn = np.array(nn)
print(nn.shape)
fig, axs = plt.subplots(2, 16, figsize=(15,8))
for i in range(2):
    for j in range(16):
        axs[i, j].imshow(nn[i * 16 + j], cmap='gray_r')
plt.show()
```
