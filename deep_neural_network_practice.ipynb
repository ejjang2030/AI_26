```python
(fashion_train_input, fashion_train_target), (fashion_test_input, fashion_test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
draw_images_100(fashion_train_input)
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(300, activation='relu'),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(fashion_train_input, fashion_train_target, epochs=30, validation_data=(fashion_test_input, fashion_test_target))
```

```python
import pandas as pd

pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.show()
```

```python
model.evaluate(fashion_test_input, fashion_test_target)
```
